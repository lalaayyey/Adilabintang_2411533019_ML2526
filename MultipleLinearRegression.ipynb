```python
from sklearn.datasets import fetch_california_housing
housing = fetch_california_housing()
df = pd.DataFrame(data= housing.data, columns=housing.feature_names)
df['MedHouseValue'] = pd.Series(housing.target) #Dependent Variable

inputs=df.drop(['MedHouseValue'],axis=1) #Indenpendent Variable
df.tail() #Menampilkan 5 data teratas
```

```python
X_train, X_test, y_train, y_test = train_test_split(inputs, df['MedHouseValue'], test_size =0.3, random_state =1)
print(X_train.shape,X_test.shape)
print(y_train.shape,y_test.shape)
```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
print(X_test)
```

```python
lr_model =LinearRegression()
lr_model.fit(X_train,y_train)
print(f'intercept :{lr_model.intercept_}')
print(f'slope : {lr_model.coef_}')
```

```python
y_pred=lr_model.predict(X_test)
mse=mean_squared_error(y_test,y_pred)
print(f'MSE : {mse}')
r2=lr_model.score(X_test,y_test)
print(f'R2 : {r2}')
plt.scatter(y_test,y_pred)
plt.xlabel('y ')
plt.ylabel('y_pred')
plt.title('Actual vs Predicted Values')
plt.show()
```
