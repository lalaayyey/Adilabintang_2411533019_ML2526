```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
```

```python
dataset_url = "https://raw.githubusercontent.com/lalaayyey/Adilabintang_2411533019_ML2526/refs/heads/main/praktikum2/Breast_cancer_data.csv"
```

```python
data = pd.read_csv(dataset_url)
print("Dataset berhasil dimuat!")
print(data.head())
```

```python
x = data.drop('diagnosis', axis=1)
y = data['diagnosis']

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.30, random_state=42
)
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
# Model dengan scaling
logreg_scaled = LogisticRegression(random_state=42)
logreg_scaled.fit(x_train_scaled, y_train)

# Model tanpa scaling
logreg_unscaled = LogisticRegression(max_iter=1000, random_state=42)
logreg_unscaled.fit(x_train, y_train)

# Ambil probabilitas untuk threshold tuning
y_prob_scaled = logreg_scaled.predict_proba(x_test_scaled)[:, 1]
y_prob_unscaled = logreg_unscaled.predict_proba(x_test)[:, 1]
```

```python
def evaluate(y_true, y_prob, threshold, label):
    y_pred = (y_prob >= threshold).astype(int)
    print(f"\n=== {label} (Threshold = {threshold}) ===")
    print(f"Accuracy : {accuracy_score(y_true, y_pred):.4f}")
    print("Classification Report:")
    print(classification_report(y_true, y_pred, target_names=['Benign (0)', 'Malignant (1)']))
```

```python
evaluate(y_test, y_prob_unscaled, 0.5, "MODEL TANPA SCALING")
evaluate(y_test, y_prob_scaled, 0.5, "MODEL DENGAN SCALING (Default)")
evaluate(y_test, y_prob_scaled, 0.3, "MODEL DENGAN SCALING (Threshold 0.3)")
evaluate(y_test, y_prob_scaled, 0.7, "MODEL DENGAN SCALING (Threshold 0.7)")
```
